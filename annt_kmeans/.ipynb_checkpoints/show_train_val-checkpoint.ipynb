```python
import pandas as pd
from matplotlib import pyplot as plt
import os
```

```python
df = pd.read_csv("results/es512_hs128_bs1_sdr10_k0of1/learning_curve_train.csv")
```

```python
df.groupby("epoch").mean().index
```

```python
plt.scatter(df["epoch"].values, df["loss"].values, color="b")
plt.xlim(0, 50)
plt.show()
```

```python
def show(path):
    if not os.path.isdir("results/" + path):
        return None
    plt.clf()
    df_train = pd.read_csv("results/%s/learning_curve_train.csv" % path).groupby("epoch").mean()
    df_val = pd.read_csv("results/%s/learning_curve_val.csv" % path)
    plt.scatter(df_train.index, df_train["loss"].values, color="b")
    plt.scatter(df_val["epoch"].values, df_val["loss"].values, color="g")
    plt.xlim(0, 50)
    plt.title(path)
    plt.show()
```

```python
show("es512_hs128_bs1_sdr10_k0of1")
```

```python
for file in os.listdir("results"):
    show(file)
```
