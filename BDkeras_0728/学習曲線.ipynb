```python
import pickle
from matplotlib import pyplot as plt
```

```python
def visualize(fp):
    data = pickle.load(open(fp,"rb"))
    print("cluster number:",data["numOfClusters"])
    print("accuracy",data["accuracy"])
    for i,history in enumerate(data["history"]):
        x = list(range(len(history["loss"])))
        plt.figure()
        plt.xlabel("number of epoch")
        plt.ylabel("value of loss / mse")
        plt.title("learning curve of cluster "+str(i))
        for label,value in history.items():
            plt.plot(x[:100], value[:100], label=label)
        #plt.plot(x, history['val_loss'], label='val_loss')
        #plt.plot(x, history['val_mean_squared_error'], label='val_mean_squared_error')
        plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        plt.show()
```

```python
visualize("./models/test1.pickle")
```

```python
visualize("./models/test10.pickle")
```

```python
visualize("./models/test2.pickle")
```
